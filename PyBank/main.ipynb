```python
# Check the data type of each column
df.info()
```

```python
# Convert the 'Date' column to datetime format
df['Date']= pd.to_datetime(df['Date'])
```

```python
# Check the format of the 'Date' column
df.info()
```

```python
# Extract month from a column in datetime format and create a dedicated column of "Month" 
# https://www.codegrepper.com/code-examples/python/how+many+months+in+the+date+in+pandas+dataframe
df['Month'] = pd.DatetimeIndex(df['Date']).month
```

```python
# Display the dataframe
display(df)
```

```python
# Use the sum function to find the net total amount of profits and losses
# https://stackoverflow.com/questions/62397627/add-a-total-row-to-a-pandas-dataframe
df = df.append(df.sum(numeric_only=True), ignore_index=True)
print(df)
```

```python
# Net total of profits and losses = $38,382,578.00.
```

```python
display(df)
```

```python
# Add new column (i.e. Diff) that shows the difference between each row in Profit/Losses 
# https://stackoverflow.com/questions/63677254/add-column-in-pandas-with-difference-of-2-rows?noredirect=1&lq=1
df['Diff'] = df['Profit/Losses'] - df['Profit/Losses'].shift(1)
```

```python
display(df)
```

```python
# Find the maximum value in the difference column to find the greatest increase in Profits over the entire period 
# https://www.kite.com/python/answers/how-to-find-the-max-value-of-a-pandas-dataframe-column-in-python
column = df["Diff"]
max_value = column.max()
print(max_value)
```
